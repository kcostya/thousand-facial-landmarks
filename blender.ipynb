```python
import os
import pandas as pd
```

```python
for _, _, files in os.walk("submits"):
    sorted(files)
```

```python
files = sorted(files)
print(files)
```

```python
submit_1 = pd.read_csv(f"submits/{files[0]}", index_col="file_name")
submit_2 = pd.read_csv(f"submits/{files[1]}", index_col="file_name")

submit_1 += submit_2
submit_1 /= 2

submit_1 = submit_1.astype(int)
submit_1.head()
```

```python
submit_1.to_csv("submit_blend.csv")
```
